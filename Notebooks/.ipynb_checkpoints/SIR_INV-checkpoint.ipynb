```python
import numpy as np
from lmfit import minimize, Parameters
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import modelos_epidemiologicos as model 
```

```python
def solve_SIR(params, t, N, I0, R0):
    beta = params['beta'].value
    gamma = params['gamma'].value
    S0 = N - I0 - R0
    y0 = S0, I0, R0
    sol = odeint(model.SIR, y0, t, args=(N, beta, gamma))
    return sol[:, 1]  # Retorna apenas I(t)
    
def residual(params, t, data, N, I0, R0):
    model = solve_SIR(params, t, N, I0, R0)
    return model - data
```

```python
# estimador I
I_data = np.array([3,7,22,	78,	233,	300,	256,	233,	189,	128,	72,	33,	11,	6])
t = np.arange(len(I_data))  # Dias
N = 763                     # População total (ajuste conforme necessário)
I0 = I_data[0]              # Infectados iniciais
R0 = 0                      # Recuperados inciasi

# Parâmetros iniciais para beta e gamma
params = Parameters()
params.add('beta', value=0.2, min=0, max=3)  # Valores iniciais e limites
params.add('gamma', value=0.1, min=0, max=1)

# Minimizar a função de resíduos
result = minimize(residual, params, args=(t, I_data, N, I0, R0))

beta_estimated = result.params['beta'].value
gamma_estimated = result.params['gamma'].value

print(f"Beta estimado: {beta_estimated}")
print(f"Gamma estimado: {gamma_estimated}")

fitted_I = solve_SIR(result.params, t, N, I0, R0)
plt.plot(t, I_data, 'o', label='Dados reais')
plt.plot(t, fitted_I, '-', label='Modelo ajustado')
```
